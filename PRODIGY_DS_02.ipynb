```python
import plotly.express as px

np.around(df.corr()[df.corr()>0.0],3)
corr = df.corr()
corr = np.around(corr[corr > 0.0],3)
fig = px.imshow(corr, text_auto=True, template='simple_white', color_continuous_scale=px.colors.sequential.Viridis, aspect="auto")
fig.update_xaxes(side="top")
fig.show()
```
